```python
def numeric_data():
    numeric_data={}
    for sentence in sent_tokens:
        numeric_data[sentence]=0
        word_tokens=nltk.word_tokenize(sentence)
        for word in word_tokens:
            if word.isdigit():     #or isbold,isunderline,isupper
                numeric_data[sentence]+=1
    maximum=max(numeric_data.values())
    for key in numeric_data:
        try:
            numeric_data[key]=numeric_data[key]/maximum
        except ZeroDivisionError:
            x=0
    print("\nnumeric_data.values=",numeric_data.values())
#     print(len(numeric_data))
    return numeric_data
```

```python
def sentence_len_score():
    sent_len_score={}
    for sentence in sent_tokens:
        word_tokens=nltk.word_tokenize(sentence)
        if len(word_tokens) in range(0,10):
            sent_len_score[sentence]=1-0.05*(10-len(word_tokens))
        elif len(word_tokens) in range(10,20):
            sent_len_score[sentence]=1
        else:
            sent_len_score[sentence]=1-0.05*(len(word_tokens)-20)
    print("\nsentence_length.values=",sent_len_score.values())
    return sent_len_score
```

```python
def sentence_position():
    sent_position={}
    n=1
    N=len(sent_tokens)
    for sentence in sent_tokens:
        a=1/n
        b=1/(N-n+1)
        sent_position[sentence]=max(a,b)
        n=n+1
    print("\nsent_position.values=",sent_position.values())
    return sent_position
```

```python
def upper_case():
    upper_case={}
    for sentence in sent_tokens:
        upper_case[sentence]=0
        word_tokens=nltk.word_tokenize(sentence)
        for word in word_tokens:
            if word.isupper():     #or isbold,isunderline,isupper
                upper_case[sentence]+=1

    maximum=max(upper_case.values())
    for key in upper_case:
        try:
            upper_case[key]=upper_case[key]/maximum
        except ZeroDivisionError:
            x=0
    print("\nupper_case.values=",upper_case.values())
    return upper_case
```

```python
def proper_noun():
    proper_noun={}
    for sentence in sent_tokens:
        proper_noun[sentence]=0
        tagged_sentence=pos_tag(sentence.split())
        proper_nouns=[word for word,pos in tagged_sentence if pos=="NNP"]   #proper noun
        proper_noun[sentence]=len(proper_nouns)
        #print(tagged_sentence)
    print(proper_noun.values())
    #get it between 0 and 1
    maximum=max(proper_noun.values())
    for key in proper_noun:
        try:
            proper_noun[key]=proper_noun[key]/maximum
        except ZeroDivisionError:
            x=0
    print("\nproper_noun.values=",proper_noun.values())
    return proper_noun
```
